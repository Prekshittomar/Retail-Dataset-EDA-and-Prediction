```python
df.drop(['CustomerID','ProductID','TransactionDate','StoreLocation','PaymentMethod'],axis=1,inplace=True)
```

```python
df['ProductCategory'].unique()
```

```python
one_hot_encoded = pd.get_dummies(df['ProductCategory'])
df_encoded = pd.concat([df, one_hot_encoded], axis=1)
```

```python
df_encoded.drop('ProductCategory',axis=1,inplace=True)
```

```python
df_encoded.head(10)
```

```python
df_encoded[['Books', 'Clothing', 'Electronics', 'Home Decor']] = df_encoded[['Books', 'Clothing', 'Electronics', 'Home Decor']].astype(int)
df_encoded
```

```python
x = df_encoded.drop('Price',axis=1)
```

```python
x.head()
```

```python
y = df_encoded['Price']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
linear = LinearRegression()
Decision = DecisionTreeRegressor()
Random = RandomForestRegressor()
```

```python
linear.fit(xtrain,ytrain)
Decision.fit(xtrain,ytrain)
Random.fit(xtrain,ytrain)
```

```python
print(linear.score(xtrain,ytrain))
```

```python
Decision.score(xtrain,ytrain)
```

```python
Random.score(xtrain,ytrain)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, median_absolute_error
import matplotlib.pyplot as plt
```

```python
linear_predictions_test = linear.predict(xtest)

linear_mae_test = mean_absolute_error(ytest, linear_predictions_test)
linear_mse_test = mean_squared_error(ytest, linear_predictions_test)
linear_rmse_test = np.sqrt(linear_mse_test)
linear_r2_test = r2_score(ytest, linear_predictions_test)
linear_median_ae_test = median_absolute_error(ytest, linear_predictions_test)

print("Linear Regression Test MAE:", linear_mae_test)
print("Linear Regression Test MSE:", linear_mse_test)
print("Linear Regression Test RMSE:", linear_rmse_test)
print("Linear Regression Test R2:", linear_r2_test)
print("Linear Regression Test Median Absolute Error:", linear_median_ae_test)
```

```python
decision_predictions_test = Decision.predict(xtest)

decision_mae_test = mean_absolute_error(ytest, decision_predictions_test)
decision_mse_test = mean_squared_error(ytest, decision_predictions_test)
decision_rmse_test = np.sqrt(decision_mse_test)
decision_r2_test = r2_score(ytest, decision_predictions_test)
decision_median_ae_test = median_absolute_error(ytest, decision_predictions_test)

print("Decision Tree Regression Test MAE:", decision_mae_test)
print("Decision Tree Regression Test MSE:", decision_mse_test)
print("Decision Tree Regression Test RMSE:", decision_rmse_test)
print("Decision Tree Regression Test R2:", decision_r2_test)
print("Decision Tree Regression Test Median Absolute Error:", decision_median_ae_test)
```

```python
random_predictions_test = Random.predict(xtest)

random_mae_test = mean_absolute_error(ytest, random_predictions_test)
random_mse_test = mean_squared_error(ytest, random_predictions_test)
random_rmse_test = np.sqrt(random_mse_test)
random_r2_test = r2_score(ytest, random_predictions_test)
random_median_ae_test = median_absolute_error(ytest, random_predictions_test)

print("Random Forest Regression Test MAE:", random_mae_test)
print("Random Forest Regression Test MSE:", random_mse_test)
print("Random Forest Regression Test RMSE:", random_rmse_test)
print("Random Forest Regression Test R2:", random_r2_test)
print("Random Forest Regression Test Median Absolute Error:", random_median_ae_test)
```

```python
import matplotlib
matplotlib.use('TkAgg')  # or 'Qt5Agg'
```

```python
models = ['Linear Regression', 'Decision Tree', 'Random Forest']
r2_scores = [linear_r2_test, decision_r2_test, random_r2_test]
plt.figure(figsize=(10, 6))
plt.bar(models, r2_scores, color=['blue', 'green', 'red'])
plt.title('R-squared Scores of Regression Models')
plt.xlabel('Models')
plt.ylabel('R-squared Score')
#plt.ylim(0, 1) 
plt.show()
```
